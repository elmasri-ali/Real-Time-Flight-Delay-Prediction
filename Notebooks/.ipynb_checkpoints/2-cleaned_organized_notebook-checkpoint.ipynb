## 1. JSON Normalization

```python
import pandas as pd
import json
```

```python
# Let's load the departure JSON data from a file named 'flight_data.json'
file_path = './data/flight_data.json'

with open(file_path, 'r') as file:
    data = json.load(file)


# Normalize the JSON data into a flat table
departure_df = pd.json_normalize(data)
```

```python
# Convert date-time fields to proper datetime objects for consistency
departure_df['arrival.scheduledTime'] = pd.to_datetime(departure_df['arrival.scheduledTime'], errors='coerce')
departure_df['arrival.estimatedTime'] = pd.to_datetime(departure_df['arrival.estimatedTime'], errors='coerce')

# Convert date-time fields to proper datetime objects for consistency
departure_df['departure.scheduledTime'] = pd.to_datetime(departure_df['departure.scheduledTime'], errors='coerce')
departure_df['departure.estimatedTime'] = pd.to_datetime(departure_df['departure.estimatedTime'], errors='coerce')
departure_df['departure.actualTime'] = pd.to_datetime(departure_df['departure.actualTime'], errors='coerce')
departure_df['departure.scheduledTime'] = pd.to_datetime(departure_df['departure.scheduledTime'], errors='coerce')
departure_df['departure.estimatedRunway'] = pd.to_datetime(departure_df['departure.estimatedRunway'], errors='coerce')
departure_df['departure.actualRunway'] = pd.to_datetime(departure_df['departure.actualRunway'], errors='coerce')
```

```python
# Create a new column for the date only
departure_df['date'] = departure_df['departure.scheduledTime'].dt.date

# Display the updated DataFrame
print(departure_df[['departure.scheduledTime', 'date']])
```

```python
departure_df.to_csv('./data/departure.csv', index=False)
```

## 2. OpenWeather Historical Data Merge

```python
import pandas as pd
import warnings
from dateutil import parser

# Ignore future warnings
warnings.simplefilter(action='ignore', category=FutureWarning)

# Load the datasets
departure_file_path = './data/departure.csv'
historical_weather_file_path = './data/openweather.csv'

departure_data = pd.read_csv(departure_file_path)
historical_weather_data = pd.read_csv(historical_weather_file_path)

# Convert timestamps to datetime format
departure_data['departure.actualTime'] = pd.to_datetime(departure_data['departure.actualTime'], errors='coerce')

# Function to parse datetime with specific format
def parse_datetime_with_format(date_str):
    try:
        return pd.to_datetime(date_str, format='%Y-%m-%d %H:%M:%S +0000 UTC')
    except parser.ParserError:
        return pd.NaT

# Apply the function to convert the dt_iso column
historical_weather_data['dt_iso'] = historical_weather_data['dt_iso'].apply(parse_datetime_with_format)

# Ensure all numeric columns are treated as numeric data types
numeric_columns = ['temp', 'pressure', 'humidity', 'visibility', 'wind_speed', 'wind_deg', 'clouds_all']
categorical_columns = ['weather_main', 'weather_description']
all_columns = ['dt_iso'] + numeric_columns + categorical_columns
historical_weather_data_filtered = historical_weather_data[all_columns]

# Drop duplicates and sort by datetime
historical_weather_data_filtered = historical_weather_data_filtered.drop_duplicates(subset=['dt_iso']).sort_values(by='dt_iso')

# Function to find the closest weather record for a given departure time
def find_closest_weather_record(departure_time, weather_data):
    time_diff = (weather_data['dt_iso'] - departure_time).abs()
    closest_time = weather_data.iloc[time_diff.argsort().iloc[:1]].iloc[0]
    return closest_time

# Apply the function to each actual departure timestamp in the departure dataset
closest_weather_data = departure_data['departure.actualTime'].apply(
    lambda ts: find_closest_weather_record(ts, historical_weather_data_filtered)
)

# Combine the original departure data with the closest weather data
combined_data = departure_data.copy()
combined_data = pd.concat([combined_data.reset_index(drop=True), closest_weather_data.reset_index(drop=True)], axis=1)

# Save the combined dataset to a new CSV file
output_file_path = './data/departure_weather_data.csv'
combined_data.to_csv(output_file_path, index=False)
```

```python
# Checking for NaN and missing values in the dataset
missing_values = df.isna().sum()
missing_values
```

```python
# Removing rows with missing values in critical columns
critical_columns = ['departure.scheduledTime', 'arrival.scheduledTime', 'airline.iataCode', 'flight.iataNumber']
df_cleaned = df.copy().dropna(subset=critical_columns)

# Fill missing values in numerical columns with the median
numerical_columns = df_cleaned.select_dtypes(include=['float64', 'int64']).columns
df_cleaned[numerical_columns] = df_cleaned[numerical_columns].fillna(df_cleaned[numerical_columns].median())

# Fill missing values in categorical columns with the mode
categorical_columns = df_cleaned.select_dtypes(include=['object']).columns
df_cleaned[categorical_columns] = df_cleaned[categorical_columns].apply(lambda x: x.fillna(x.mode()[0]))

# Verify that there are no missing values left
missing_values_cleaned = df_cleaned.isna().sum()
missing_values_cleaned
```

All missing values have been handled in the dataset. The following strategies were applied:

Rows with missing values in critical columns (departure.scheduledTime, arrival.scheduledTime, airline.iataCode, and flight.iataNumber) were removed. Numerical columns had their missing values filled with the median. Categorical columns had their missing values filled with the mode

```python
# Plotting box plots for numerical columns to visually inspect for outliers
plt.figure(figsize=(20, 10))

numerical_columns = ['departure.terminal', 'departure.delay', 'temp', 'pressure', 'humidity', 'visibility', 'wind_speed', 'wind_deg', 'clouds_all']
for i, column in enumerate(numerical_columns, 1):
    plt.subplot(3, 3, i)
    sns.boxplot(y=df_cleaned[column])
    plt.title(f'Box Plot of {column}')
    plt.grid(True)

plt.tight_layout()
plt.show()
```

```python
# Function to calculate outliers using adjusted IQR method (Q1=0.2, Q3=0.8)
def find_outliers_adjusted_iqr(data, lower_quantile=0.2, upper_quantile=0.8):
    Q1 = data.quantile(lower_quantile)
    Q3 = data.quantile(upper_quantile)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    outliers = data[(data < lower_bound) | (data > upper_bound)]
    return outliers

# Identifying outliers for each numerical column using adjusted IQR method
adjusted_outliers_dict = {}
for column in numerical_columns:
    adjusted_outliers_dict[column] = find_outliers_adjusted_iqr(df_cleaned[column])

# Display the number of outliers in each column using adjusted thresholds
adjusted_outliers_count = {column: len(outliers) for column, outliers in adjusted_outliers_dict.items()}
print("Number of outliers in each column using adjusted IQR method:", adjusted_outliers_count)

# Removing outliers based on the adjusted IQR method
for column in numerical_columns:
    outliers = adjusted_outliers_dict[column]
    df_cleaned = df_cleaned[~df_cleaned.index.isin(outliers.index)]

# Verify the new shape of the dataset
print("Original shape:", df.shape)
print("New shape after removing outliers:", df_cleaned.shape)

# Plotting box plots for numerical columns after filtering
plt.figure(figsize=(20, 10))
for i, column in enumerate(numerical_columns, 1):
    plt.subplot(3, 3, i)
    sns.boxplot(y=df_cleaned[column])
    plt.title(f'Box Plot of {column} after Removing Outliers')
    plt.grid(True)

plt.tight_layout()
plt.show()
```
